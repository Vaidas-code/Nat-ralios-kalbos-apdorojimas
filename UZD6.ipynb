
#### 1. Pateiktam sakiniui pateikite priklausomybių gramatikos medį, pakeisdami jo fono ir teksto spalvas (spalvas parinkite savo nuožiūra). Atspausdinkite mazgus ir priklausomybių žymas.

```python
import spacy
from spacy import displacy

tekstas = 'The students exam scores were all fantastic'

nlp = spacy.load("en_core_web_sm")

doc = nlp(tekstas)

for token in doc:
    print(f"Žodis: {token.text}, Priklausomybė: {token.dep_}, Mazgas: {token.head.text}")

pasirinkimai = {
    "bg": "green",
    "color": "black",
    "font": "Times New Roman"
}

displacy.render(doc, style="dep", options=pasirinkimai, jupyter=True)
```

#### 2. Tam pačiam sakiniui pateikite priklausomybių gramatikos medį parinkdami kompaktišką vaizdavimo režimą bei atspausdindami lemas atskiroje eilutėje po sakinio tokenais.

```python
import spacy
from spacy import displacy

tekstas ='The students exam scores were all fantastic'

nlp = spacy.load("en_core_web_sm")

doc = nlp(tekstas)

for token in doc:
    print(f"Žodis: {token.text:{10}}, Priklausomybė: {token.dep_:{10}}, Mazgas: {token.head.text}")
print("\n")
for token in doc:
    print(f"Lema: {token.lemma_}")

pasirinkimai = {
    "compact": True,
    "bg": "orange",
    "color": "black",
    "font": "Times New Roman"
}
displacy.render(doc, style="dep", options=pasirinkimai, jupyter=True)
```

#### 4. Pateiktam sakiniui pateikite priklausomybių gramatikos medį, pakeisdami jo fono iš teksto spalvas (spalvas parinkite savo nuožiūra). Atskiroje eilutėje po sakinio tokenais atspausdinkite lemas. Atspausdinkite mazgus ir priklausomybių žymas.¶

```python
import spacy
from spacy import displacy

nlp = spacy.load("lt_core_news_sm")

tekstas = "Žmogus be mokslo kaip namai be langų"

doc = nlp(tekstas)

options = {
    "compact": True,
    "bg": "green",
    "color": "black",
    "font": "Arial",
}

displacy.render(doc, style="dep", options=options, jupyter=True)
    
print("\n")

for token in doc:
    print(f"{token.text} -> {token.lemma_}")

print("\nMazgai ir priklausomybių žymos:")
for token in doc:
    print(f"Žodis: {token.text}, Priklausomybė: {token.dep_}, Mazgas: {token.head.text}")
```

#### 5. Iš duoto tekstinio failo "Tekstas.txt" atrinkite tik tuos sakinius, kuriuose yra veiksmažodžiai. Kiekvienam sakiniui, kuriame yra veiksmažodis, pavaizduokite gramatinį priklausomybių medį (dependency tree).

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

with open("Tekstas.txt", "r", encoding="utf-8") as file:
    tekstas = file.read()

doc = nlp(tekstas)

def has_verb(sentence):
    for token in sentence:
        if token.pos_ == "VERB":
            return True
    return False

for sent in doc.sents:
    if has_verb(sent):
        print(f"Sakinys: {sent.text}")
        displacy.render(sent, style="dep", jupyter=True)


```

#### 6. Nustatykite kiekvieno sakinio galvą (angl. head) ir jos vaikus (angl. children)

```python
import spacy

nlp = spacy.load("lt_core_news_sm")

tekstas = "Vaikai žaidžia parke. Jis skaito naują knygą. Saulė šviečia ryškiai."

doc = nlp(tekstas)

for sent in doc.sents:
    print(f"\nSakinys: {sent.text}")
    for token in sent:
        if token == token.head:
            print(f"Galva: {token.text}")
            children = [child.text for child in token.children]
            print(f"Vaikai: {', '.join(children)}")
```
